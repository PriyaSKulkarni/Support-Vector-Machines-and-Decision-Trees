```python
type(train_data)
x_train= train_data.iloc[:, 0:4]
y_train = train_data.iloc[:,4:]
print(x_train)
print(y_train)
x_train = x_train.to_numpy()
print(type(x_train))
print(x_train)
y_train = y_train.to_numpy()
print(type(y_train))
print(y_train)
```

```python
# Just giving numbers to the class so that it is easy to predict
for i in range(len(y_train)):
    if y_train[i] == ' Plastic ':
        y_train[i] = 1
    elif y_train[i] == ' Non-Plastic ':
        y_train[i] = 2
```

```python
y_train
```

```python
#prediction accuracy function 
def prediction_Accuracy(x_test,y_pred,y_test):
  predit_sum = 0
  for i in range(len(x_test)):
    if(y_pred[i] == y_test[i]):
      predit_sum = predit_sum + 1
  return (predit_sum/len(x_test))*100
```

```python
# Importing the sklearn SVM library
from sklearn.svm import SVC
x_train=x_train.astype('float64')
x_test=x_test.astype('float64')
y_train=y_train.astype('float64')
y_test=y_test.astype('float64')
```

Using the linear SVM wth regularization term C =10

```python
linear_clf =  SVC(kernel = 'linear',C=10)
linear_clf.fit(x_train, y_train)  # Train the algorithm using the training set data
y_pred_train= linear_clf.predict(x_train) # predicting the output for the training set 
y_pred_train=np.array(y_pred_train)
print(y_pred_train) # predicted output
print("Accuracy on training data :",prediction_Accuracy(x_train,y_pred_train,y_train)) #Accuracy of the training set
```

```python
y_pred_test= linear_clf.predict(x_test) # predicting the output for test data
y_pred_test=np.array(y_pred_test)
print("predicted output",y_pred_test) # predicted output of test data
print("Accuracy on test data :", prediction_Accuracy(x_test,y_pred_test,y_test)) # the accuracy for the test data
```

I think the linear SVM is not overfitting  as it is giving an prediction accuracy of 72.22% 

```python
# importing the requried data to plot the datapoints and decision boundary
import seaborn as sns
import matplotlib.pyplot as plt
```

Plotting the datapoints and decision boundary in a 2D projections for features (ignoring hue feature) as follows :
1. height and diameter
2. diameter and weight
3. height and weight

Note: the support vectors are encircled in each plot

```python
# Using 2D projections for plotting
# plotting Height and diameter features values
# plotting the training set values
sns.scatterplot(x=x_train[:,0], y=x_train[:,1], hue=y_pred_train.flatten(),s=20);
#plotting the testing set values
sns.scatterplot(x=x_test[:,0], y=x_test[:,1], hue=y_pred_test.flatten(),s=20, palette=['green','orange']);
# Constructing a hyperplane using a formula. 
w = linear_clf.coef_[0]           # w consists of 2 elements
b = linear_clf.intercept_[0]      # b consists of 1 element
x_= np.linspace(-0.0001,-0.17)    # generating x points 
y_ = (-(w[0] / w[1]) * x_) - b / w[1]  # getting corresponding y points
plt.plot(x_, y_, c='r');        # Plotting a red hyperplane
plt.scatter(linear_clf.support_vectors_[:, 0],linear_clf.support_vectors_[:, 1], s=50, 
           facecolors='none',  edgecolors='k', alpha=.5); # support vectors are circled
w_ = w[0:2] / (np.sqrt(np.sum(w[0:2] ** 2))) # w hat value
margin_value = 1 / np.sqrt(np.sum(w[0:2] ** 2)) # margin value
decision_boundary = np.array(list(zip(x_, y_)))
points_above = decision_boundary + w_ * margin_value
points_below = decision_boundary - w_ * margin_value 
# plotting the margin above and below
plt.plot(points_above[:, 0], points_above[:, 1], 'b--', linewidth=2) 
plt.plot(points_below[:, 0], points_below[:, 1], 'g--', linewidth=2)
```

```python
# Using 2D projections for plotting
# plotting diameter and weight features values
# plotting the training set values
sns.scatterplot(x=x_train[:,1], y=x_train[:,2], hue=y_pred_train.flatten(),s=20);
#plotting the testing set values
sns.scatterplot(x=x_test[:,1], y=x_test[:,2], hue=y_pred_test.flatten(),s=20, palette=['green','orange']);
# Constructing a hyperplane using a formula.
w = linear_clf.coef_[0]           # w consists of 2 elements
b = linear_clf.intercept_[0]      # b consists of 1 element
x_ =np.linspace(0.0003, 0.35)    # generating x points 
y_ = -(w[1] / w[2]) * x_ - b / w[2]  # getting corresponding y points
plt.plot(x_, y_, c='r');      # Plotting a red hyperplane
plt.scatter(linear_clf.support_vectors_[:, 1],linear_clf.support_vectors_[:, 2], s=50, facecolors='none', 
            edgecolors='k', alpha=.5); # support vectors are circled
w_ = w[1:3] / (np.sqrt(np.sum(w[1:3] ** 2)))  # w hat value
margin_value = 1 / np.sqrt(np.sum(w[1:3] ** 2)) # margin value
decision_boundary = np.array(list(zip(x_, y_)))
points_above = decision_boundary + w_ * margin_value
points_below = decision_boundary - w_ * margin_value
# plotting the margin above and below
plt.plot(points_above[:, 0], points_above[:, 1], 'b--', linewidth=2)
plt.plot(points_below[:, 0], points_below[:, 1], 'g--', linewidth=2)
```

Using SVM for Guassian kernel:

Note : Since rbf kernel is same a gaussian kernel in sklearn library, i am using it

```python
# using the sklearn library kernel = gaussiann
# the gamma value manipulates the sigma value and gamma value used here is 60
# the regularization term is C = 10
gaussian_clf =  SVC(kernel = 'rbf',gamma=60, C=10)
gaussian_clf.fit(x_train, y_train) # training the algorithm with the training dataset
```

```python
# predicting the output for training data
y_pred_train= gaussian_clf.predict(x_train)
y_pred_train =np.array(y_pred_train)
print("predicted output :",y_pred_train)
print("Accuracy on train data :", prediction_Accuracy(x_train,y_pred_train,y_train)) # the accuracy of the training data
```

```python
#predicting the output for test data
y_pred_test= gaussian_clf.predict(x_test)
y_pred_test =np.array(y_pred_test)
print("predicted output :",y_pred_test)
print("Accuracy on test data :", prediction_Accuracy(x_test,y_pred_test,y_test)) # the accuracy of the test data
```

there is no overfitting happening in gaussian kernel SVM as we can see that the accuracy on test data is 83.34% which is very good

Comparing the results of linear and gaussian kernel SVM, the gaussian kernel svm performs better than the linear svm because the gaussian kernel svm gives an accuracy rate of 83.34% on test data whereas the linear svm gives an accuracy rate of 72.22% which is comparetively less than the gaussian kernel svm.

we can conclude that the data given is not linearly separable.
